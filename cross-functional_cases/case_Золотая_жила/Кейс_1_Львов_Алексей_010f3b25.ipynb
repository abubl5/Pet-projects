```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df_original = pd.read_csv('golden_mine1.csv')
df_cleaned = df_original.copy()
```

```python
#Визуально осмотрим данные на наличие выбросов по столбцам 'Payments' и 'Costs' и построим график 
#Также уберем единичные выбросы в ручную по индексам строк

#Проверим на отсутствие пропусков
df_c = df_cleaned.copy()
display(df_c.isna().sum())
```

```python
#Построим box plot для визуализации выбросов
plt.figure(figsize=(12, 5))

#Box plot для Payments
plt.subplot(1, 2, 1)
plt.boxplot(df_cleaned['Payments'], whis=2)
plt.title('Box Plot: Payments')
plt.ylabel('Значение платежей (руб.)')
plt.grid(True)

#Box plot для Costs
plt.subplot(1, 2, 2)
plt.boxplot(df_cleaned['Costs'], whis=2)
plt.title('Box Plot: Costs')
plt.ylabel('Значение затрат (руб.)')
plt.grid(True)

#Отступы для выделения выбросов
plt.subplot(1, 2, 1)
plt.margins(y=0.1)
plt.subplot(1, 2, 2)
plt.margins(y=0.1)

plt.tight_layout()
plt.show()

#Проверим min для каждого столбца, чтобы убедиться в отсутствии отрицательных значений
print("Минимальное значение Payments:", df_cleaned['Payments'].min())
print("Минимальное значение Costs:", df_cleaned['Costs'].min())
```

```python
#Уберем выбросы
#Payments
display(df_original.sort_values(by='Payments').head(10))
display(df_original.sort_values(by='Payments', ascending=False).head(10))
df_cleaned = df_cleaned.drop(index = 8322)
display(df_cleaned.sort_values(by='Payments', ascending=False).head(10))
```

```python
#Costs
display(df_original.sort_values(by='Costs').head(10))
display(df_original.sort_values(by='Costs', ascending=False).head(10))
df_cleaned = df_cleaned.drop(index = 2759)
display(df_cleaned.sort_values(by='Costs', ascending=False).head(10))
```

Я заметил, что в колонке Age есть разные форматы записи гендера, приведем все к одному виду

```python
df_cleaned['Gender'] = df_cleaned['Gender'].str.capitalize()
df_cleaned['Gender'].unique()
```

**Приведем данные о профите после очистки**

```python
#Кол-во клиентов после очистки данных от выбросов
count_people = len(df_cleaned)
display(count_people)

#Прибыль с одного клиента на текущий момент
profit_1 = df_cleaned['Profit'].mean()
print(f'Прибыль с привлечения одного клиента: {profit_1}')

#Общая прибыль c привлечения 1000 клиентов без учета таргетирования на основе стреднего значения
full_profit = (profit_1 - 15000)*1000
print(f'Общая прибыль с привлечения без таргетинга: {full_profit}')
```

**Разделим по категориям и посмотрим общий профит каждой категории**

!!!Как я понял, так как я аналитик и смотрю на потенциальную аудиторию, то я должен производить расчет с учетом вычета затрат на привлечение (15000)

```python
display(df_cleaned['AgeGroup'].unique())

#Категория Female 18-24
df_female18_24 = df_cleaned[(df_cleaned['AgeGroup'] == '18-24') & (df_cleaned['Gender'] == 'Female')]
profit_female18_24 = df_female18_24['Profit'].mean() - 15000
print(f'Прибыль с привлечения одного клиента категории Female 18-24: {profit_female18_24}')
full_profit_female18_24 = profit_female18_24*1000
print(f'Общая прибыль с привлечения 1000 клиентов категории Female 18-24: {full_profit_female18_24}')
```

```python
 #Категория Male 18-24
df_male18_24 = df_cleaned[(df_cleaned['AgeGroup'] == '18-24') & (df_cleaned['Gender'] == 'Male')]
profit_male18_24 = df_male18_24['Profit'].mean() - 15000
print(f'Прибыль с привлечения одного клиента категории Male 18-24: {profit_male18_24}')
full_profit_male18_24 = profit_male18_24*1000
print(f'Общая прибыль с привлечения 1000 клиентов категории Male 18-24: {full_profit_male18_24}')
```

**Из этих двух категорий Male 18-24 и Female 18-24 я бы выбрал категорию Female 18-24, т.к. у них значительно больше профит, относительно среднего значения уже с учетом цены привлечения в 15000**

Пройдемся по всем остальным группам

```python
#Female 25-39
df_female25_39 = df_cleaned[(df_cleaned['AgeGroup'] == '25-39') & (df_cleaned['Gender'] == 'Female')]
profit_female25_39 = df_female25_39['Profit'].mean() - 15000
print(f'Прибыль с привлечения одного клиента категории Female 25-39: {profit_female25_39.astype(int)}')
full_profit_female25_39 = profit_female25_39*1000
print(f'Общая прибыль с привлечения 1000 клиентов категории Female 25-39: {full_profit_female25_39.astype(int)}')
print()

#Male 25-39
df_male25_39 = df_cleaned[(df_cleaned['AgeGroup'] == '25-39') & (df_cleaned['Gender'] == 'Male')]
profit_male25_39 = df_male25_39['Profit'].mean() - 15000
print(f'Прибыль с привлечения одного клиента категории Male 25-39: {profit_male25_39.astype(int)}')
full_profit_male25_39 = profit_male25_39*1000
print(f'Общая прибыль с привлечения 1000 клиентов категории Male 25-39: {full_profit_male25_39.astype(int)}')
print()

#Female 40
df_female40 = df_cleaned[(df_cleaned['AgeGroup'] == '40+') & (df_cleaned['Gender'] == 'Female')]
profit_female40 = df_female40['Profit'].mean() - 15000
print(f'Прибыль с привлечения одного клиента категории Female 40+: {profit_female40.astype(int)}')
full_profit_female40 = profit_female40*1000
print(f'Общая прибыль с привлечения 1000 клиентов категории Female 40+: {full_profit_female40.astype(int)}')
print()

#Male 40
df_male40 = df_cleaned[(df_cleaned['AgeGroup'] == '40+') & (df_cleaned['Gender'] == 'Male')]
profit_male40 = df_male40['Profit'].mean() - 15000
print(f'Прибыль с привлечения одного клиента категории Male 40+: {profit_male40.astype(int)}')
full_profit_male40 = profit_male40*1000
print(f'Общая прибыль с привлечения 1000 клиентов категории Male 40+: {full_profit_male40.astype(int)}')
```

```python
# Визуализация профита по категориям (раздельные столбики для каждого пола и возраста)

categories = [
    "Female 18-24", "Male 18-24",
    "Female 25-39", "Male 25-39",
    "Female 40+", "Male 40+"
]
profits = [
    profit_female18_24,
    profit_male18_24,
    profit_female25_39,
    profit_male25_39,
    profit_female40,
    profit_male40
]

plt.figure(figsize=(10, 6))
plt.bar(categories, profits, color=["pink", "lightblue", "pink", "lightblue", "pink", "lightblue"])
plt.title("Профит с привлечения одного клиента с учетом затрат на привлечение")
plt.ylabel("Профит в рублях")
plt.xlabel("Половозрастная категория")
plt.grid(axis='y')
plt.xticks(rotation=30)
plt.show()
```
